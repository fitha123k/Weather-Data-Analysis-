* **Precipitation (`precip_mm`):** A strong right-skew indicates that most observations experience little to no rainfall, while heavy precipitation events occur rarely.

* **Humidity:** The distribution peaks at higher values (around 80–90%), indicating that humid conditions are common across the dataset.

* **Cloud cover (`cloud`):** The multimodal pattern suggests discrete reporting levels (e.g., 0, 20, 40, etc.), indicating categorized cloud coverage rather than continuous measurement.

* **Visibility (`visibility_km`):** A strong concentration near higher values (around 10 km) indicates that clear visibility conditions are dominant, with poor visibility being uncommon.

* **Carbon monoxide (`co`):** Values are concentrated near zero with a long tail, indicating that CO levels are generally low, with occasional elevated readings.

* **Ozone (`o3`):** Most values fall within a moderate range, with some higher outliers indicating episodic increases influenced by environmental conditions.

* **Nitrogen dioxide (`no2`):** The distribution is heavily skewed toward low values, suggesting that NO₂ levels are typically low but can spike under certain conditions.

* **Sulfur dioxide (`so2`):** Values remain close to zero for most observations, indicating minimal SO₂ presence and limited contribution to overall pollution.

```python
df_grouped = (
    df_new.groupby('condition_text', as_index=False)['temp_c']
    .mean()
    .sort_values('temp_c')
)

# Plot
plt.figure(figsize=(10, 15))
sns.barplot(
    data=df_grouped,
    x='temp_c',
    y='condition_text',
    color='teal'
)

plt.title('Average Temperature by Weather Condition')
plt.xlabel('Average Temperature (°C)')
plt.ylabel('Condition_text')
plt.show()
```

### INFERENCE:

- Coldest averages are linked to snowy and icy conditions (patchy snow, blizzard, freezing drizzle).  
- Mid-range temperatures align with fog, drizzle, light rain, and overcast conditions.  
- Warmer averages correspond to clear, sunny, partly cloudy, and rain-heavy conditions.  
- Thunderstorm-related weather (heavy rain with thunder, patchy light rain with thunder) appears at the warmest end.  
- Overall, the plot shows a clear gradient: freezing precipitation at the low end, progressing upward through rain and cloudiness, peaking with sunny and stormy weather.
